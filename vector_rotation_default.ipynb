# Default Word2Vec Embedding Variations

We're interested in the default variation between two embeddings generated by `word2vec` on the same dataset using the same parameters. These vectors are trained using `gensim`.

````python
%matplotlib inline
from lib.w2v.w2v import *
````

````python
model_1 = model_from_saved("./temp/models-ec2/newsgroup", binary=False)
model_2 = model_from_saved("./temp/models-clic/newsgroup", binary=False)
````

These models are trained with the following parameters

```
min_count=5, workers=8, iter=300, window=15, size=300, negative=25
```

Users from Columbia would recognize why I named the models so `:p`

Since these two models are from the same data set, they'd share the same vocabulary. Then, we can calculate the cosine similarity of the vectors of all the words in the vocabulary from the two models.

````python
from scipy import spatial

vocabs = {}

for word in model_1.vocab:
    if word in model_1 and word in model_2:
        cosine_sim = 1 - spatial.distance.cosine(model_1[word], model_2[word])
        vocabs[word] = cosine_sim

import matplotlib.pyplot as plot

plot.hist(vocabs.values(), bins=100)
plot.figure()
````

Turns out the rotation is not a constant value, but a value somewhat normally distributed around `0+`. There's also a tiny blip around 1.0, where the vectors are (curiously) exactly the same (since we're measuring cosine similarity). This result holds constant for other data sets.

Let's verify this result on another dataset.

````python
model_3 = model_from_saved("./temp/models-ec2/enron", binary=False)
model_4 = model_from_saved("./temp/models-clic/enron", binary=False)

from scipy import spatial

vocabs = {}

for word in model_3.vocab:
    if word in model_3 and word in model_4:
        cosine_sim = 1 - spatial.distance.cosine(model_3[word], model_4[word])
        vocabs[word] = cosine_sim

import matplotlib.pyplot as plot

plot.hist(vocabs.values(), bins=100)
plot.figure()
````
